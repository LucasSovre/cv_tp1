```python
import numpy as np
import math
```

```python
np.random.seed(1)
```

```python
X = np.random.rand(1000) * 3

mean = round(np.mean(a=X),3)
std = round(np.std(X),3)
median = round(np.median(X),3)
```

```python
X_bis = np.random.rand(1000) * 3

mean_bis = round(np.mean(X_bis),3)
std_bis = round(np.std(X_bis),3)
median_bis = round(np.median(X_bis),3)
```

```python
print(f"Les differences sont : \n mean : {mean} , {mean_bis} \n ecrat type : {std} {std_bis} \n median : {median} {median_bis}")
```

```python
noise = np.random.randn(1000) * 0.1
X = np.random.rand(1000) * 3
y = np.sin(X) + noise
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X, y)
plt.show()
```
